Цель: изучения возможностей визуализации алгоритмов

Функция: np.exp(-x)+np.sin(5*x)-10

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
from IPython.display import display


def step_slice(lst, step):
    return lst[step]


# параметры анимации
def animate_list(lst, play=False, interval=200):
    slider = widgets.IntSlider(min=0, max=len(lst) - 1, step=1, value=0)
    if play:
        play_widjet = widgets.Play(interval=interval)
        widgets.jslink((play_widjet, "value"), (slider, "value"))
        display(play_widjet)
        # slider = widgets.Box([play_widject, slider])
    return interact(step_slice, lst=fixed(lst), step=slider)


limx = [-6, 10]


def f(x):
    return np.exp(-x) + np.sin(5 * x) - 10  # Заданная вариантом функция


# Отрисовка
def plot_trajectory(func, x_l, x_r, x_p, flag_end, limit_point=None):
    fig = plt.figure(figsize=(5, 5))  # Создаем рисунок и задаем его размеры
    ax = fig.add_axes([0, 0, 2, 1])  # Задаем маштаб
    # ax.grid(True) # включить сетку графика
    delta = 0.5  # шаг между точкам
    x = np.arange(limx[0], limx[1], delta)  # Вектор оси x
    ax.plot(x, f(x), "k")
    ax.plot([np.min(x), np.max(x)], [0, 0], "--", color="black")  # точка границы
    ax.grid()
    ax.set_title("Метод хорд")
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    for i in range(len(x_l) - 1):
        ax.plot([x_l[i], x_l[i]], [0, f(x_l[i])], "--r", alpha=0.1)  # левая граница
        ax.plot(x_l[i], f(x_l[i]), "*", color="black", alpha=0.1)  # точка границы
        ax.plot([x_r[i], x_r[i]], [0, f(x_r[i])], "--r", alpha=0.1)  # правая граница
        ax.plot(x_r[i], f(x_r[i]), "*", color="black", alpha=0.1)  # точка границы
        ax.plot(x_p[i], f(x_p[i]), "*", color="r", alpha=0.1)  # точка среднего значения
        ax.plot([x_l[i], x_r[i]], [f(x_l[i]), f(x_r[i])], "--r", alpha=0.1)
    # i=-1
    ax.plot([x_l[-1], x_l[-1]], [0, f(x_l[-1])], "--g")  # левая граница
    ax.plot(x_l[-1], f(x_l[-1]), "*", color="black")  # точка границы
    ax.plot([x_r[-1], x_r[-1]], [0, f(x_r[-1])], "--g")  # правая граница
    ax.plot(x_r[-1], f(x_r[-1]), "*", color="black")  # точка границы
    ax.plot(x_p[-1], f(x_p[-1]), "*", color="g")  # точка среднего значения
    if flag_end == 1:
        ax.annotate(
            "Корень= %.2f" % x_p[-1],
            xy=(x_p[-1], f(x_p[-1])),
            xycoords="data",
            xytext=(2, 40),
            textcoords="data",
            size=20,
            va="center",
            ha="center",
            arrowprops=dict(arrowstyle="simple", connectionstyle="arc3,rad=0.2"),
        )
    plt.close(fig)
    return fig


# начальные условия
x_l = limx[0] + 1
x_r = limx[1] - 1

x_l_ = np.array([x_l])
x_r_ = np.array([x_r])
x_p_ = np.array([])
plots = []
e = 0.01
flag_end = 0
# тела вычислений
while abs(x_l - x_r) > e:
    print(x_l, x_r, abs(f(x_r)))
    x_p = x_r - ((f(x_r) * (x_r - x_l)) / (f(x_r) - f(x_l)))
    # Формируем динамически добавляющиеся вектора
    x_l_ = np.append([x_l_], [x_l])
    x_r_ = np.append([x_r_], [x_r])
    x_p_ = np.append([x_p_], [x_p])
    plots.append(plot_trajectory(f, x_l_, x_r_, x_p_, flag_end, limit_point=(0, 0)))

    if f(x_p) * f(x_l) < 0:
        x_r = x_p
    else:
        x_l = x_p
    # Последняя итерация
    if abs(f(x_r)) < e:
        flag_end = 1
        plots.append(plot_trajectory(f, x_l_, x_r_, x_p_, flag_end, limit_point=(0, 0)))
        break

animate_list(plots, play=True, interval=200)
```
