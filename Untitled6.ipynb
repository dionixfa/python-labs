```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

def f(x):
    # Здесь задайте вашу функцию
    return x**3 - 5*x**2 + 2*x + 1

def chord_method(x0, x1, tol=1e-5, max_iter=100):
    """
    Метод хорд для нахождения корня уравнения f(x) = 0.

    Параметры:
        x0, x1 (float): Начальные точки для метода хорд.
        tol (float): Точность (по умолчанию 1e-5).
        max_iter (int): Максимальное количество итераций (по умолчанию 100).

    Возвращает:
        float: Приближенное значение корня уравнения f(x) = 0.
    """
    for i in range(max_iter):
        x_next = x1 - f(x1) * (x1 - x0) / (f(x1) - f(x0))
        if abs(x_next - x1) < tol:
            return x_next
        x0, x1 = x1, x_next
    raise ValueError("Метод не сошелся за заданное количество итераций")

# Визуализация анимации
x_vals = np.linspace(-2, 6, 500)
y_vals = f(x_vals)

fig, ax = plt.subplots()
ax.plot(x_vals, y_vals, label='f(x)')
ax.axhline(y=0, color='black', linewidth=0.5)
ax.axvline(x=0, color='black', linewidth=0.5)
ax.grid()

# Начальные точки для метода хорд
x0 = 1.0
x1 = 3.0

line, = ax.plot([], [], 'ro-', label='Текущая точка')
iteration_text = ax.text(0.02, 0.95, '', transform=ax.transAxes)

def init():
    line.set_data([], [])
    iteration_text.set_text('')
    return line, iteration_text

def update(frame):
    global x0, x1

    # Ограничиваем количество итераций для анимации
    if frame < 20:
        x_next = ((x1 - f(x1) * (x1 - x0)) / (f(x1) - f(x0)))
        line.set_data([x0, x1, x_next], [f(x0), f(x1), f(x_next)])
        x0, x1 = x1, x_next
        iteration_text.set_text(f'Итерация: {frame}')

    return line, iteration_text

ani = FuncAnimation(fig, update, frames=np.arange(0, 20), init_func=init, blit=True)
plt.legend()
from IPython.display import HTML, display
display(HTML(ani.to_jshtml()))
plt.close()
```
